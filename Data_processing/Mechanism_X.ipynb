```python
import boto3

aws_access_key_id = '********'
aws_secret_access_key = '*************'
region_name = 'ap-south-1'  # e.g., Mumbai

s3 = boto3.client('s3',
                  aws_access_key_id=aws_access_key_id,
                  aws_secret_access_key=aws_secret_access_key,
                  region_name=region_name)
bucket_name = 'bank-transactions-pipeline'
s3_prefix = 'txn_chunks/'
```

```python
import pandas as pd
import os
from time import sleep

# Setup
csv_path = '/content/drive/MyDrive/transactions_processed.csv'  # Change this
chunk_size = 10000
chunk_number = 0
os.makedirs('/content/chunks_temp', exist_ok=True)

def upload_chunk_to_s3(chunk, chunk_number):
    filename = f"chunk_{chunk_number:05}.parquet"
    local_path = f"/content/chunks_temp/{filename}"
    s3_key = f"{s3_prefix}{filename}"

    # Save and upload
    chunk.to_parquet(local_path, index=False)
    s3.upload_file(local_path, bucket_name, s3_key)
    print(f"✅ Uploaded: {s3_key}")
    os.remove(local_path)

# Read and upload chunks
for chunk in pd.read_csv(csv_path, chunksize=chunk_size):
    chunk_number += 1
    upload_chunk_to_s3(chunk, chunk_number)
    sleep(1)  # wait 1 second per chunk

print("🎉 All chunks uploaded to S3.")
```
